```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baseline_score = '/Users/julian/temp/BAYESIAN_SKIP_SAVED_EXPERIMENTS/ISLES/std_isles_baseline_prior_as_input/loss_log.xlsx'
conv_bayes_score = '/Users/julian/temp/BAYESIAN_SKIP_SAVED_EXPERIMENTS/ISLES/std_isles_convolution_bayesian_skip/loss_log.xlsx'
# baseline_score = '/Users/julian/temp/BAYESIAN_SKIP_SAVED_EXPERIMENTS/GSD/all_maps_with_ncct_with_core_unet_pct_multi_att_dsv/loss_log_1596961662.xlsx'
# conv_bayes_score = '/Users/julian/temp/BAYESIAN_SKIP_SAVED_EXPERIMENTS/GSD/bayesian_skip_connection/loss_log.xlsx'
```

```python
train_baseline_df = pd.read_excel(baseline_score)
train_conv_bayes_df = pd.read_excel(conv_bayes_score)
train_conv_bayes_df.head()
```

```python
train_loss_fig = plt.figure(figsize=(5, 5))
ax = train_loss_fig.add_subplot(1,1,1)

train_baseline_df.rolling(5).mean().plot(ax=ax,x='epoch', y='Seg_Loss', label='Unet', grid=True, style=':')
train_conv_bayes_df.rolling(5).mean().plot(ax=ax, x='epoch', y='Seg_Loss', label='Unet + Method B', grid=True, style='--')
ax.set_xlabel('Epochs')
ax.set_ylabel('Train Loss')
```

```python
train_loss_fig.savefig('../static/figures/train_loss.eps', format='eps')
```

```python
baseline_df = pd.read_excel(baseline_score, sheet_name='test')
conv_bayes_df = pd.read_excel(conv_bayes_score, sheet_name='test')
conv_bayes_df.head()
```

```python
loss_fig = plt.figure(figsize=(5, 5))
ax = loss_fig.add_subplot(1,1,1)

baseline_df.rolling(5).mean().plot(ax=ax, x='epoch', y='Seg_Loss', label='Unet', grid=True, style=':')
conv_bayes_df.rolling(5).mean().plot(ax=ax, x='epoch', y='Seg_Loss', label='Unet + Method B', grid=True, style='--')
ax.set_xlabel('Epochs')
ax.set_ylabel('Validation Loss')
```

```python
loss_fig.savefig('../static/figures/validation_loss.eps', format='eps')


```
