```python
import sys
sys.path.insert(0, '../')
import numpy as np
import torch
from dataio.transformation.imageTransformations import RandomFlipTransform
from gsprep.visual_tools.visual import display
```

```python
dataset_path = '/Users/julian/stroke_research/data/all2016_subset_prepro/data_set.npz'
channel = 0
subj = 1
raw_images = np.load(dataset_path, allow_pickle=True)['ct_inputs'][subj][..., channel].astype(np.float64)
# raw_labels = np.load(dataset_path, allow_pickle=True)['ct_lesion_GT'][subj].astype(np.uint8)
raw_labels = np.load(dataset_path, allow_pickle=True)['lesion_GT'][subj].astype(np.float64)
raw_images.shape
```

```python
display(raw_images)
display(raw_labels)
```

```python
flip_axis = (2)
random_flip_prob = 1
seed = 42
max_output_channels = 2
```

```python
transform = RandomFlipTransform(axes=flip_axis, flip_probability=1, p=random_flip_prob, seed=seed, max_output_channels=max_output_channels)
images, labels = torch.from_numpy(raw_images), torch.from_numpy(raw_labels)
transformed_input, transformed_label = transform(images, labels)
```
